## Jogos do Brasileirão (2013-2017)


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as sns
```

Importando os dataframes arenas, clubes, estados e jogos(2013-2017) e indicando quais serão os índices

```python
arenas = pd.read_csv("/Users/tulioteodoro/Documents/GitProjects/brasileirao/data/arenas.csv", index_col=['id_arena'])
clubes = pd.read_csv("/Users/tulioteodoro/Documents/GitProjects/brasileirao/data/clubes.csv", index_col=['id_clube'])
estados = pd.read_csv("/Users/tulioteodoro/Documents/GitProjects/brasileirao/data/estados.csv", index_col=['id_estado'])
jogos2013 = pd.read_csv("/Users/tulioteodoro/Documents/GitProjects/brasileirao/data/jogos_2013.csv", index_col=['id_arena', 'id_clube1', 'id_clube2', 'id_clube_vencedor', 'id_estado_vencedor', 'id_estado_clube1', 'id_estado_clube2'])
jogos2014 = pd.read_csv("/Users/tulioteodoro/Documents/GitProjects/brasileirao/data/jogos_2014.csv", index_col=['id_arena', 'id_clube1', 'id_clube2', 'id_clube_vencedor', 'id_estado_vencedor', 'id_estado_clube1', 'id_estado_clube2'])
jogos2015 = pd.read_csv("/Users/tulioteodoro/Documents/GitProjects/brasileirao/data/jogos_2015.csv", index_col=['id_arena', 'id_clube1', 'id_clube2', 'id_clube_vencedor', 'id_estado_vencedor', 'id_estado_clube1', 'id_estado_clube2'])
jogos2016 = pd.read_csv("/Users/tulioteodoro/Documents/GitProjects/brasileirao/data/jogos_2016.csv", index_col=['id_arena', 'id_clube1', 'id_clube2', 'id_clube_vencedor', 'id_estado_vencedor', 'id_estado_clube1', 'id_estado_clube2'])
jogos2017 = pd.read_csv("/Users/tulioteodoro/Documents/GitProjects/brasileirao/data/jogos_2017.csv", index_col=['id_arena', 'id_clube1', 'id_clube2', 'id_clube_vencedor', 'id_estado_vencedor', 'id_estado_clube1', 'id_estado_clube2'])
```

```python
jogos2013.head(2)
```

```python
#Concatenando os jogos de 2013 a 2017 em somente um df.

jogos = pd.concat([jogos2013, jogos2014, jogos2015, jogos2016, jogos2017], axis=0)
jogos.head()
```

Para se trabalhar com os dados o ideal é que tenhamos um dataframe que possua todos os dados necessários para a manipulação de acordo com o que buscamos extrair de informação.

```python
#Join de jogos com as arenas
jogos = jogos.join(arenas, how='left', on=['id_arena'])
jogos.head()
```

```python
clubes.head(1)
```

```python
#Join dos jogos com os clubes
jogos = jogos.join(clubes, on=['id_clube1'], lsuffix='1', rsuffix='2')
jogos = jogos.join(clubes, on=['id_clube2'], lsuffix='1', rsuffix='2')
jogos.head()
```

```python
jogos = jogos.join(clubes, on=['id_clube_vencedor'])
jogos.head()
```

```python
#Renomeando o nome da coluna desc_clube que no join recebeu o nome original 
jogos.rename(columns={'desc_clube':'desc_clube_vencedor'}, inplace=True)
```

```python
estados.head(2)
```

```python
jogos.head(2)
```

```python
#join entre os jogos e os nomes dos clubes
jogos = jogos.join(estados, on=['id_estado_clube1'], lsuffix='1', rsuffix='2')
jogos = jogos.join(estados, on=['id_estado_clube2'], lsuffix='1', rsuffix='2')
jogos.head()
```

```python
#Junção entre os jogos e os nomes dos Estados vencedores
jogos = jogos.join(estados, on=['id_estado_vencedor'])
jogos.rename(columns={'desc_estado':'desc_estado_vencedor', 'sigla':'sigla_vencedor'}, inplace=True)
jogos.head()
```
